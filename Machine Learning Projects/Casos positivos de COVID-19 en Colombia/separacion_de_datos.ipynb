```python
import pandas as pd
```

```python
data = pd.read_csv('Datasets/covid_col_limpio.csv')
data.info()
```

```python
columnas = ['Nombre departamento', 'Nombre municipio', 'Sexo', 'Fecha inicio sintomas', 'Fecha diagnostico','Edad categorizada', 'Tipo de contagio', 'Ubicacion caso', 'Estado']
for col in columnas:
    data[col] = data[col].astype("category")
```

```python
data.info()
```

```python
data.head()
```

```python
data = data[['Nombre departamento', 'Nombre municipio', 'Edad', 'Edad categorizada', 'Sexo', 'Fecha diagnostico', 'Fecha inicio sintomas','Tipo de contagio', 'Ubicacion caso', 'Estado']]
```

```python
data.head()
```

## Separando departamentos

```python
departamentos = data[['Nombre departamento']]
departamentos.head(-10)
```

```python
departamentos_unicos = departamentos['Nombre departamento'].unique()
departamentos = pd.DataFrame(departamentos_unicos, columns=['Nombre dertamento'])
departamentos
```

## Separando municipios

```python
municipios = data[['Nombre municipio']]
municipios_unicos = municipios['Nombre municipio'].unique()
municipios = pd.DataFrame(municipios_unicos, columns=['Nombre municipio'])
municipios
```

## Separando Fechas

```python
fechas = data[['Fecha inicio sintomas']]
fechas_unicas = fechas['Fecha inicio sintomas'].unique()
fechas = pd.DataFrame(fechas_unicas, columns=['Fecha inicio sintomas'])
fechas
```

```python
departamentos.to_csv("Datasets/departamentos.csv")
municipios.to_csv("Datasets/municipios.csv")
fechas.to_csv("Datasets/fechas.csv")
```
