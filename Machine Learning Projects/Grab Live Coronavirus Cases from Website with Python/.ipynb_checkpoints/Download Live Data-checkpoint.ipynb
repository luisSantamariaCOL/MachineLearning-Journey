# Download Live Data

## Importing libraries

```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

## Testing the URL and sending request to grab live data

```python
URL = "https://www.worldometers.info/coronavirus/#countries"

r = requests.get(URL)
html_elements = r.text
```

```python
html_elements
```

## BeautifulSoup for extracting table data

```python
soup = BeautifulSoup(html_elements, 'lxml')
```

```python
soup
```

```python
get_table = soup.find("table", id = "main_table_countries_today")
get_table
```

```python
get_table_data = get_table.tbody.find_all("tr")
get_table_data
```

```python
dic = {}
for i in range(len(get_table_data)):
    try:
        key = get_table_data[i].find_all("a", href = True)[0].string
    except:
        continue
    values = [j.string for j in get_table_data[i].find_all('td')]
    values.pop(7)
    dic[key] = values
```

## Convert data to DataFram

```python
pd.DataFrame(dic)
```

```python
df = pd.DataFrame(dic).iloc[1:,:].T.iloc[:,1:8]
df
```

```python
column_names = ['Total Cases', 'New Cases', 'Total Deaths', 'New Deaths', 'Total Recovered',
               'Active Cases', 'Serious, Critical']
df.index_name = "country"
df.columns = column_names
```
