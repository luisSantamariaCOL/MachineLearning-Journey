```python
import numpy as np

a = np.arange(25).reshape(5,5)
a
```

```python
# red selection
red = a[:, 1::2]
red
```

```python
# yellow selection
yellow = a[-1,:]
yellow
```

```python
# blue selection
pre_blue = a[1::2, :3:2]
print(pre_blue)
```

```python
pre_blue = a[1::2, :4:2]
print(pre_blue)
```

```python
pre_blue = a[1::2, :-1:2]
print(pre_blue)
```
