# matplotlib

Matplotlib es una **librería** para la generación de gráficos a partir de datos contenidos en listas o arrays en el lenguaje de programación python y su extensión matemática NumPy.

```python
from matplotlib import pyplot as plt
```

```python
# Vamos a añadir nuestros propios datos
# Creando una lista
# Creando una lista de valores para el eje x del gráfico.

# Promedio de salario de Developers por edad
dev_x = [x for x in range(25,36)]

dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
```

```python
print('dev_x =', dev_x)
print('dev_y =', dev_y)
```

```python
%matplotlib inline

# Creamos una gráfica de 2 dimensiones x y
plt.plot(dev_x, dev_y)

# La graficamos
plt.show()
```
