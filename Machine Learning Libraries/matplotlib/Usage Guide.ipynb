```python
import matplotlib.pyplot as plt
import numpy as np
```

## A simple example
Matplotlib graphs your data on Figures. Each of which can obtain one or more **Axes**.
The most simple way of creating a figure with an axes is using pyplot.subplots.

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,4,2,3])
```

```python
fig = plt.figure()
```

```python
fig, ax = plt.subplots()
```

```python
fig, axs = plt.subplots(2,2)
```

```python
x = np.linspace(0,2,100)

# Note that even in the 00-style, we use '.pyplot,figure' to create the figure.
fig, ax = plt.subplots() # Create a figure and an axes.
ax.plot(x, x, label='linear') # Plot some data on axes.
ax.plot(x, x**2, label='quadratic') # Plot more data on the axes...
ax.plot(x, x**3, label='cubic') # ... and some more.
ax.set_xlabel('x label') # Add an x-label to the axes.
ax.set_ylabel('y label') # Add a y-label to the axes.
ax.legend() # Add a legend
```
