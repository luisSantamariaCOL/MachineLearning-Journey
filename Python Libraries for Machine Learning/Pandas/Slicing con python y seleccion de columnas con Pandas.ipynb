### Slicing

#### cadena[inicio:fin]

```python
variable = "pollitos"
slicing = variable[2:7]
print(slicing)
```

```python
# Desde el índice 2 hasta el final
slicing = variable[2:]
print(slicing)
```

```python
# Desde el inicio hasta el índice 3 excluyente
slicing = variable[:3]
print(slicing)
```

```python
# Desde el índice final -1 excluyente hasta el índice -6
slicing = variable[-6:-1]
print(slicing)
```

#### slicing con saltos cadena[inicio:fin:salto]

```python
# 2 saltos cada letra
slicing = variable[::2]
print(slicing)
```

### Selección con iloc: Filas, columnas y combinación de filas  columnas 

![](https://shanelynnwebsite-mid9n9g1q9y8tt.netdna-ssl.com/wp-content/uploads/2016/10/Pandas-selections-and-indexing.png)

```python
# all rows all columns
df.iloc[:,:]
```

```python
# first value of dataframe, in index 0 row and index 0 column
df.iloc[0,0]
```

```python
# last value of dataframe, in last index row and last index column. Type numpy.float64
df.iloc[-1,-1]
```

```python
# first 3 rows of dataframe, and all the columns
df.iloc[0:3,:]
```

```python
# same first 3 rows of dataframe, and all the columns
df.iloc[:3, :]
```

```python
# all rows and first 3 columns of dataframe.
df.iloc[:,:3]
```

```python
# rows with id 0, 1 and 2, and all the columns.
df.iloc[[0,1,2],:]
```

```python
# rows 0, 2 and 4. columns 0, 5 and 6
df.iloc[[0,2,4], [0,5,6]]
```

```python
# value in the row index 1 and column index 5. Type String
df.iloc[1,5]
```

```python
# first 5 rows, first 5 columns
df.iloc[0:5, 0:5]
```

```python
# same first 5 rows, and first 5 columns
df.iloc[:5, :5]
```

```python
# First row. Type Series
df.iloc[0]
```

```python
# row index 1, value in the column with index 5. Type Series
df.iloc[1:2, 5]
```

```python
# row index 1, value in the column with index 5. Type DataFrame
df.iloc[1:2, [5]]
```
